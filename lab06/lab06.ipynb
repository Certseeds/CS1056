```python
import matplotlib.pyplot as plt
# Set the size of figure
plt.figure(figsize=(8,4))
# Prepare the data
x = [1,2,3,4,5,6,7,8]
y = [1,2,3,4,5,6,7,8]
# Plot the data
plt.plot(x, y, 'go',label='dots')
plt.legend()# Show legend
plt.title('Title') # Add title
plt.xlabel('X')    # Add x label
plt.ylabel('Y')    # Add y label
#save figure
plt.savefig('pre_dots.png', dpi=300)
# Show the plot
plt.show()
```

```python
#Exercise 1
import pandas as pd
import numpy as np
gas=pd.read_csv("./gas_prices.csv",low_memory=False)
gas
x  = gas["Year"]
y1 = gas["Canada"]
y2 = gas["France"]
y3 = gas["Germany"]

# Plot the data
plt.xticks(ticks=np.arange(1990,2011,step=3))
plt.plot(x, y1,color='blue',marker='o',label='Canada')
plt.plot(x, y2,color='orange',marker='+',label='France')
plt.plot(x, y3,color='green',marker='*',label='Germany')

plt.legend()# Show legend
plt.title('Gas Prices over Time') # Add title
plt.xlabel('Year')    # Add x label
plt.ylabel('US Dollars')    # Add y label
#save figure
#plt.savefig('pre_dots.png', dpi=300)
# Show the plot
plt.show()
```

```python
#exercise 2
import random
top=pd.read_csv("./top_10_population_2020.csv",low_memory=False)
top
top_countrys = top["Country (or dependency)"]
top_popus = top["Population (2020)"]
plt.barh(top_countrys, top_popus, height=0.8,color=[(random.random(), random.random(), random.random()) for x in range(10)])
plt.legend()# Show legend
plt.title('top_10_population_2020') # Add title
plt.xlabel('population')    # Add x label
plt.ylabel('country name')    # Add y label
plt.text(s='129millions',x=top_popus[0]/10,y=9)
#save figure
#plt.savefig('pre_dots.png', dpi=300)
# Show the plot
plt.show()
```
