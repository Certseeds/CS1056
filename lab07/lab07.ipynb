```python
import matplotlib.pyplot as plt
import pandas as pd
import random as random
SSD = pd.read_csv("./SleepStudyData.csv")
print(SSD['Hours'].value_counts())

plt.bar(["no","yes"],SSD["Enough"].value_counts(),color=[(random.random(), random.random(), random.random()) for x in range(10)])
plt.legend()# Show legend
plt.title('top_10_population_2020') # Add title
plt.xlabel('population')    # Add x label
plt.ylabel('country name')    # Add y label

plt.show()
#save figure
#plt.savefig('pre_dots.png', dpi=300)
# Show the plot
temp=SSD.groupby("Tired")
plt.bar(temp.groups.keys(),temp["Hours"].median(),color=[(random.random(), random.random(), random.random()) for x in range(10)])
plt.legend()# Show legend
plt.title('top_10_population_2020') # Add title
plt.xlabel(';')    # Add x label
plt.ylabel('country name')    # Add y label
```

```python
import numpy as np
import nibabel as nib
print(12)
def read_nil(file_path:str):
    ct_scan = nib.load(file_path)
    array = ct_scan.get_fdata()
    array = np.rot90(np.array(array))
    return array

sample_ct = read_nil("./CT/ct_scans/coronacases_org_001.nii")
sample_ct1 = read_nil("./CT/infection_mask/coronacases_001.nii")
sample_ct2 = read_nil("./CT/lung_and_infection_mask/coronacases_001.nii")
sample_ct3 = read_nil("./CT/lung_mask/coronacases_001.nii")
list1 = [sample_ct,sample_ct1,sample_ct2,sample_ct3]

for i in range(3,4,1):
    plt.subplot(1,4)
    plt.imshow(list1[i][...,150],cmap="bone")
    plt.imshow(list1[0][...,150],cmap='bone')
```
