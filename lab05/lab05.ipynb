```python
import matplotlib.pyplot as plt
# Set the size of figure
plt.figure(figsize=(8,4))
# Prepare the data
x = [1,2,3,4,5,6,7,8]
y = [1,2,3,4,5,6,7,8]
# Plot the data
plt.plot(x, y, 'go',label='dots')
plt.legend()# Show legend
plt.title('Title') # Add title
plt.xlabel('X')    # Add x label
plt.ylabel('Y')    # Add y label
#save figure
plt.savefig('pre_dots.png', dpi=300)
# Show the plot
plt.show()
```

```python
#Exercise 1
import pandas as pd
import numpy as np
gas=pd.read_csv("./gas_prices.csv",low_memory=False)
gas
x  = gas["Year"]
y1 = gas["Canada"]
y2 = gas["France"]
y3 = gas["Germany"]

# Plot the data
plt.xticks(ticks=np.arange(1990,2011,step=3))
plt.plot(x, y1, 'b.-',label='Canada')
plt.plot(x, y2, 'y.-',label='France')
plt.plot(x, y3, 'r.-',label='Germany')

plt.legend()# Show legend
plt.title('Gas Prices over Time') # Add title
plt.xlabel('Year')    # Add x label
plt.ylabel('US Dollars')    # Add y label
#save figure
#plt.savefig('pre_dots.png', dpi=300)
# Show the plot
plt.show()
```

```python
#exercise 2
top=pd.read_csv("./top_10_population_2020.csv",low_memory=False)
top
top_countrys = top["Country (or dependency)"]
top_popus = top["Population (2020)"]
plt.barh(top_countrys, top_popus, height=0.8)
plt.legend()# Show legend
plt.title('top_10_population_2020') # Add title
plt.xlabel('population')    # Add x label
plt.ylabel('country name')    # Add y label
#save figure
#plt.savefig('pre_dots.png', dpi=300)
# Show the plot
plt.show()
```

```python
# this csv file please find in release
owid=pd.read_csv("./owid-covid-data.csv",low_memory=False)
owid_peoplesum=owid.groupby(by="location")["new_cases"].sum()
print(owid_peoplesum.name)
print(owid_peoplesum[0])
lines = owid_peoplesum.where(owid_peoplesum < 1000000).dropna()
owid_peoplesum = owid_peoplesum.drop(lines.index).drop("World").sort_values(ascending=True)
# print(owid_peoplesum.index)
# print(owid_peoplesum.values)
plt.bar(owid_peoplesum.index,owid_peoplesum.values,width=0.8)
plt.title('extara') # Add title
plt.xlabel('countrys')    # Add x label
plt.ylabel('peoplenumber')    # Add y label
#save figure
#plt.savefig('pre_dots.png', dpi=300)
# Show the plot
plt.show()
```
