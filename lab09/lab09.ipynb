```python
import pandas as pd 
import numpy as np 
import time 
import matplotlib.pyplot as plt
```

```python
def func(path: str) ->None:
    music = pd.read_csv(path)
    row_num, column_num = music.shape
    time_series_freq = 0.005  # 5*10^-3, 200HZ
    temp = music['TS1']
    print(music.head())
    features = ["Channel1", "Channel2", "Channel3","Channel4"]
    print(type(temp[0]))
    timeArray = []
    for i in temp:
        if(type(i) == float):
            timeArray.append(timeArray[len(timeArray)-1])
            continue
        timeStruct = time.strptime("1971-01-01 "+ i,"%Y-%m-%d %H:%M:%S")
        timeArray.append(time.mktime(timeStruct))
    for index in range(0,column_num,1):
        timeArray[index] = timeArray[index] + (index % 200) * time_series_freq 
    
    for name_of_order in features:
        x = timeArray
        y = music[name_of_order]
        y_array = y
        print(y_array.mean())
        print(y_array.max())
        print(y_array.min())
        plt.figure(num=3)
        plt.title(name_of_order)
        plt.ylim(y_array.min(),y_array.max())
        plt.plot(x, y_array, alpha=0.6, label=name_of_order)
        plt.xlabel('row')
        plt.ylabel('column')

        plt.legend()
        plt.show()
        print("123")
    temp
func("./tiny_eeg_self_experiment_music.csv")
func("./tiny_eeg_self_experiment_reading.csv")
```
